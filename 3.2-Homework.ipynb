### 3. **Create a set of about 3 elements**

* #### _Add a list of fruits to the above set and print the result_
* #### _Add a car element to your set_
* #### _Create a 2nd set with a few odd items_
* #### _Save the union of 1st set and 2nd set to a 3rd set_
* #### _Pop an element from the 2nd set, and print the remainder of the set_
* #### _Clear the 1st set and print the result_
* #### _Discard an element, and remove another element from the 3rd set_
* #### _Print the remainder of the 3rd set_

```python
# Create a set of about 3 elements
set1 = {2023, "Penguin", "Hiphop"}


# Add a list of fruits to the above set and print the result


fruitlist = ["cherries", "grapes", "melon", "dragonfruit", "orange"]

set1.update(fruitlist)

print("Updated set is ...", set1)


# Add a car element to your set

set1.add("Tire")

print("Car element added to set1 ...\n", set1) 



# Create a 2nd set with a few odd items


set2 = {"grilled cheese", "yellow", "guam"}


# Save the union of 1st set and 2nd set to a 3rd set


set3 = set1.union(set2)

print("United sets ...", set3)


# Pop an element from the 2nd set, and print the remainder of the set

set2.pop()

print("Set2 after pop is ...", set2)


# Clear the 1st set and print the result

set1.clear()

print("Set1 cleared ...", set1)


# Discard an element, and remove another element from the 3rd set


set3.discard("yellow")

set3.remove("Penguin")

# Print the remainder of the 3rd set

print("Remainder of set3 is ...", set3)


```
